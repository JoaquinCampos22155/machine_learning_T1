## Analisis de columnas


```python
selected_columns = [
    "Track",
    "Track Score",
    "Spotify Streams",
    "TikTok Posts",
    "All Time Rank"
]

print("=" * 75)
print("ANÁLISIS DE COLUMNAS SELECCIONADAS")

print("\nTipos de datos:\n")

for column in selected_columns:
    print(f"{column:<20} -> {df[column].dtype}")

print("\n" + "=" * 75)
print("VISTA PREVIA")

display(df[selected_columns].head(10))


print("\n" + "=" * 75)
print("CORRELACIÓN ENTRE VARIABLES NUMÉRICAS")

numeric_columns = [
    "Track Score",
    "Spotify Streams",
    "TikTok Posts",
    "All Time Rank"
]

correlation = df[numeric_columns].corr(numeric_only=True)

display(correlation)
```

Como entre las columnas a evaluar solo 1 es numérica no las podemos correlacionar entre ellas directamente por lo tanto

```python
columns_to_check = [
    "Spotify Streams",
    "TikTok Posts",
    "All Time Rank"
]

for col in columns_to_check:

    print("\n" + "="*60)
    print(f"Revisando columna: {col}")

    temp = df[col].astype(str).str.replace(",", "", regex=False)

    invalid = temp[pd.to_numeric(temp, errors="coerce").isna()]

    print("Cantidad de valores no numéricos:", invalid.nunique())
    print("Ejemplos:")

    print(invalid.unique()[:10])
```

ahora sabemos que las columnas "Spotify Streams", "TikTok Posts" y "All Time Rank" contienen valores numéricos del todo los unicos que no son son los NULL digamos, por lo tanto podemos proceder a hacer limpieza 

```python
for col in columns_to_check:

    original = df[col].notna().sum()

    converted = pd.to_numeric(
        df[col].astype(str).str.replace(",", "", regex=False),
        errors="coerce"
    ).notna().sum()

    print(f"{col}:")
    print(f"Valores originales: {original}")
    print(f"Valores convertibles: {converted}")
    print(f"Se perderían: {original-converted}")
    print()
```

```python
print("="*75)
print("CONVIRTIENDO COLUMNAS A Int64")

for col in columns_to_check:

    df[col] = (
        df[col]
        .astype(str)
        .str.replace(",", "", regex=False)
    )

    df[col] = pd.to_numeric(
        df[col],
        errors="coerce"
    )

print(df[columns_to_check].dtypes)
```

```python
df_clean = df.dropna(
    subset=[
        "Track Score",
        "Spotify Streams",
        "TikTok Posts",
        "All Time Rank"
    ]
)
print(df_clean.dtypes)
```

ahora si recalculamos la correlación

```python
corr = df_clean[
    [
        "Track Score",
        "Spotify Streams",
        "TikTok Posts",
        "All Time Rank"
    ]
].corr()

display(corr)
```

```python
print("=" * 75)
print("MATRIZ DE CORRELACIÓN")

plt.figure(figsize=(8,6))

sns.heatmap(
    corr,
    annot=True,        
    fmt=".2f",         
    cmap="coolwarm",   
    linewidths=0.5,
    square=True
)

plt.title("Correlation Matrix", fontsize=14)
plt.xticks(rotation=20)
plt.yticks(rotation=0)

plt.show()
```

# Descripción de las 5 columnas 

Para este análisis se seleccionaron cinco columnas que representan diferentes aspectos del desempeño y la popularidad de una canción dentro de Spotify y otras plataformas digitales, nos llamaron la atencion estas columasn para explorar si existe algún tipo de relacion directa y sobre todo explorar el valor del score y como este se ve afectado

### Track

Corresponde al nombre de la canción. Es una variable categórica (texto) utilizada para identificar cada registro dentro del dataset.

### Track Score

Es una puntuación numérica asignada a la canción según su desempeño general en el dataset. Resume diferentes indicadores de popularidad y permite comparar el rendimiento de las canciones.

### Spotify Streams

Representa la cantidad total de reproducciones que ha obtenido una canción en Spotify. Es uno de los principales indicadores de popularidad dentro de la plataforma.

### TikTok Posts

Indica el número de publicaciones en TikTok que utilizan o hacen referencia a la canción. Esta variable permite medir el impacto y la viralidad de una canción.

### All Time Rank

Corresponde al ranking histórico de la canción dentro del dataset. Un valor más pequeño representa una mejor posición en la clasificación general.

*Como evaluación nos gustaria determinar y confirmar nuestra hipótesis que es que el track score y el all time rank si tienen que estar correlacionados bastante fuerte*

```python
import matplotlib.pyplot as plt

print("=" * 70)
print("CORRELACIÓN ENTRE TRACK SCORE Y ALL TIME RANK")

corr = df_clean[["Track Score", "All Time Rank"]].corr()

display(corr)

plt.figure(figsize=(7,5))

plt.scatter(
    df_clean["Track Score"],
    df_clean["All Time Rank"],
    alpha=0.6
)

plt.title("Track Score vs All Time Rank")
plt.xlabel("Track Score")
plt.ylabel("All Time Rank")

plt.grid(True)

plt.show()
```
